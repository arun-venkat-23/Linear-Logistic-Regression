By visual inspection, we could infer that all features does not have a normal distribution. Moreover fitting a bell curve on top the plot confirms that the distribution is not normal and the data have multiple peaks. 

There are outliers present in the distribution which could be inferred from the histogram plots. Peaks/bars that are distant from the major/central peak signify the presence of outliers. The sensor values of -200 which is quite far from the rest of the data distribution are the major outliers and make the distribution non-normal.Given a sample, we could not say if they are actual outliers or some rare values that needs to be considered for training the model. For this sample of data, they act as outliers skewing the data.

Since the data has outliers and is non-normally distributed, it could bias the model(slopes and intercept) if not normalized. ALso, the given data has features with diverse ranges (different range of values). If the ranges are different, it would be difficult to compare the corresponding feature's importance. And also if not scaled or normalized, the predictions would be poor and misinterpreted. The above factors lead to suboptimal performance of the model. Thus normalization is required for the given dataset.

I employed Z-score standardization to normalize the data. It transforms the data of all features such that it has a mean of 0 and a standard deviation of 1, making the distribution close to normal distribution. This ensures that all features are in a common scale and contribute equally to the analysis. Since the skewed points might be outliers or might not, the points are not removed and are thus included while training the model. Standardization does not change the shape of the distribution, meaning if the original data is skewed the standardized data will retain those, which is what we wanted. The formula for standardization is:

$$ X_{norm} = \frac{X - \mu}{\sigma} $$

Where:
- **𝑋𝑛𝑜𝑟𝑚:** is the standardized value of \( X \),
- **X:** is the original value of the considered feature,
- **𝜇:** is the mean of the considered feature,
- **𝜎:** is the standard deviation of the considered feature.

```python
#Function to standardize the data. Data is standardized in order to being them to the same range.
def standardization_data(df):
    for cols in df.columns[2:]:
        #Computing mean of the data
        mean = df[cols].mean()
        #Computing standard deviation of the data
        std = df[cols].std()
        #Standardizing the feature, the features will have a mean = 0 and standard deviation = 1
        df[cols] = (df[cols] - mean)/std
        #Plotting histogram for all features
        df[cols].hist()
        plt.title(f"Standardized Histogram of {cols}")
        plt.xlabel(cols)
        plt.ylabel("Value")
        plt.tight_layout()
        plt.show()
        plt.tight_layout()
        plt.show()
```

```python
def pearson_coefficient(feature1, feature2):
    r, p = pearsonr(df[feature1], df[feature2])
    return r

def scatter_plot(feature1, feature2):
    #Creating the scatter plot
    plt.scatter(df[feature1], df[feature2])
    plt.title(f"Scatter Plot of {feature1} vs. {feature2}")
    plt.xlabel(feature1)
    plt.ylabel(feature2)
    plt.tight_layout()
    plt.grid(True)
    plt.show()
```

**Pearson’s Correlation coefficient** tell you how strongly two variables are related along with the proportionality (inverse or direct). The coefficient ranges from -1 to 1. <br> 
>    **+1: Perfect positive correlation** - both variables increase or decrease together perfectly. <br>
>    **−1: Perfect negative correlation** - one variable increases as the other decreases perfectly. <br>
>     **0: No linear correlation** - there’s no straight-line relationship between the variables. <br>

Yes, there is a high correlation between the two features PT08.S2(NMHC) and PT08.S5(O3). High correlation indicates that change in one feature will also induce a change in the other correlated feature. Both features move closely.

```python
#Converting from pandas to numpy
x_train_array = x_train.to_numpy()
y_train_array = y_train.to_numpy()
```

```python
def manual_linear_regression(x_train_array, y_train_array, learning_rate, num_iterations):
    sample_size = x_train_array.shape[0]
    slope = np.zeros((x_train_array.shape[1], 1))
    cost_list = []
    x_values = []
    y_values = []
    for i in range(num_iterations):
        y_prediction = np.dot(x_train_array, slope)
        cost = (1/sample_size)*np.sum(np.square(y_prediction - y_train_array))
        dcost_dslope = (1/sample_size)*np.dot(x_train_array.T, y_prediction - y_train_array)
        slope = slope - learning_rate * dcost_dslope
        cost_list.append(cost)
    return cost_list, slope, y_prediction
```

```python
cost, slope, y_prediction = manual_linear_regression(x_train_array, y_train_array, 0.01, 1000)

# Print the slope values corresponding to the feature_names
features_to_display = ['Intercept','NMHC(GT)','C6H6(GT)','PT08.S2(NMHC)','NOx(GT)','PT08.S3(NOx)','NO2(GT)','PT08.S4(NO2)','PT08.S5(O3)','T','RH','AH']
print("Predicted intercept and slope values: ")
for i, feature_name in enumerate(features_to_display):
    print(f"{feature_name}: {slope[i][0]:.6f}")
```

# Linear Regression 5-fold Cross Validation

```python
# Perform 5-fold cross-validation
kf = KFold(n_splits=5, shuffle=True, random_state=42)
rmse_list = []

for train_index, val_index in kf.split(df):
    x_train_cv = df.iloc[train_index, 2:]
    x_val = df.iloc[val_index, 2:]
    y_train_cv = df.iloc[train_index, 1:2]
    y_val = df.iloc[val_index, 1:2]
    
    # Adding Bias term
    x_train_cv.insert(0,"Bias",1)  
    x_val.insert(0,"Bias",1)

    #Converting from pandas to numpy
    x_train_array_fold = x_train_cv.to_numpy()
    x_val_fold = x_val.to_numpy()
    y_train_fold = y_train_cv.to_numpy()
    y_val_fold = y_val.to_numpy()
    
    # Train the model
    cost_list, slope, _ = manual_linear_regression(x_train_array_fold, y_train_fold, 0.01, 1000)

    # Calculate RMSE on validation set
    y_pred = np.dot(x_val_fold, slope)
    rmse = np.sqrt(np.mean(np.square(y_pred - y_val_fold)))
    rmse_list.append(rmse)
    
    # Print results
    print("RMSE for each fold:", rmse_list)
    
print("Average RMSE:", np.mean(rmse_list))
print("Standard deviation of RMSE:", np.std(rmse_list))
```

No big change was noticed across the average and standard deviation values across all 5-folds, the changes were minimal.

The coefficients indicate the impact of each feature on the target variable. Features with higher absolute coefficients (regardless of the direction - positive or negative) are considered more informative because they have a greater influence on the target variable.

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
def manual_logistic_regression(x_train_array, y_train_binary, learning_rate_logistic, num_iterations_logistic):
    epsilon = 0
    sample_size_logistic = x_train_array.shape[0]
    slope_logistic = np.zeros((x_train_array.shape[1], 1))
    cost_list_logistic = []
    for i in range(num_iterations_logistic):
        z = np.dot(x_train_array, slope_logistic)
        y_prediction = sigmoid(z)
        y_prediction_logistic = np.clip(y_prediction, 1e-15, 1 - 1e-15) 
        cost = (-1/sample_size_logistic) * np.sum((y_train_binary+epsilon)*np.log(y_prediction_logistic + epsilon) + (1-y_train_binary+epsilon)*np.log(1-y_prediction_logistic+epsilon))
        dcost_dslope = (1/sample_size_logistic)*np.dot(x_train_array.T, y_prediction_logistic - y_train_binary)
        slope_logistic = slope_logistic - learning_rate_logistic * dcost_dslope
        cost_list_logistic.append(cost)
        #if i % 10000 == 0:
            #cost = (-1/sample_size_logistic) * np.sum((y_train_binary+epsilon)*np.log(y_prediction_logistic + epsilon) + (1-y_train_binary+epsilon)*np.log(1-y_prediction_logistic+epsilon))
            #print(f'Cost after iteration {i+1}: {cost}')
    return cost_list_logistic, slope_logistic, y_prediction_logistic
```

# Logisitc Regression 5-fold Cross Validation and ROC Curve

```python
def logistic_metrics(y_pred_logistic, y_val_fold_logistic):
    # Count true positives, true negatives, false positives, and false negatives
    tp = np.sum((y_val_fold_logistic == 1) & (y_pred_logistic == 1))
    tn = np.sum((y_val_fold_logistic == 0) & (y_pred_logistic == 0))
    fp = np.sum((y_val_fold_logistic == 0) & (y_pred_logistic == 1))
    fn = np.sum((y_val_fold_logistic == 1) & (y_pred_logistic == 0))

    # Calculate accuracy, precision, recall, and F1-score
    accuracy = (tp + tn) / (tp + tn + fp + fn)
    precision = tp / (tp + fp)
    recall = tp / (tp + fn)
    f1 = 2 * (precision * recall) / (precision + recall)

    return accuracy, precision, recall, f1
```

```python
# Perform 5-fold cross-validation
kf = KFold(n_splits=5, shuffle=True, random_state=42)
accuracy_logistic = []
precision_logistic = []
recall_logistic = []
f1_logistic = []
individual_plots = []

fold_colors = ['b', 'g', 'r', 'c', 'm']

for i,(train_index, val_index) in enumerate(kf.split(df)):
    x_train_cv_logistic = df.iloc[train_index, 2:]
    x_val_logistic = df.iloc[val_index, 2:]
    y_train_cv_logistic = df.iloc[train_index, 1:2]
    y_val_logistic = df.iloc[val_index, 1:2]
    
    # Adding Bias term
    x_train_cv_logistic.insert(0,"Bias",1)  
    x_val_logistic.insert(0,"Bias",1)

    #Converting from pandas to numpy
    x_train_array_fold_logistic = x_train_cv_logistic.to_numpy()
    x_val_fold_logistic = x_val_logistic.to_numpy()
    y_train_fold_logistic = y_train_cv_logistic.to_numpy()
    y_val_fold_logistic = y_val_logistic.to_numpy()
    
    #print(x_train_array_fold_logistic.shape)
    #print(x_val_fold_logistic.shape)
    #print(y_train_fold_logistic.shape)
    #print(y_val_fold_logistic.shape)
    
    # Train the model
    _, slope_logistic_cv, _ = manual_logistic_regression(x_train_array_fold_logistic, y_train_fold_logistic, 0.01, 10000)

    # Calculate metrics on validation set
    z_logistic = np.dot(x_val_fold_logistic, slope_logistic_cv)
    y_pred_logistic = (sigmoid(z_logistic)>0.5).astype(int)
    accuracy, precision, recall, f1 = logistic_metrics(y_val_fold_logistic, y_pred_logistic)
    accuracy_logistic.append(accuracy)
    precision_logistic.append(precision)
    recall_logistic.append(recall)
    f1_logistic.append(f1)
    
    # Calculate FPR, TPR, and thresholds
    fpr, tpr, thresholds = roc_curve(y_val_fold_logistic, y_pred_logistic)
    roc_auc = auc(fpr, tpr)
    
    fig, ax = plt.subplots()

    ax.plot(fpr, tpr, label='ROC curve (area = %0.6f)' % roc_auc, color = fold_colors[i])
    ax.legend(loc="lower right")
    ax.set_xlim([-0.02, 1.02])
    ax.set_ylim([-0.02, 1.02])
    ax.set_xlabel('False Positive Rate')
    ax.set_ylabel('True Positive Rate')
    ax.set_title(f'ROC Curve for Logistic Regression {i+1} Fold')
    ax.plot([0, 1], [0, 1], color='orange', linestyle='--')
    ax.grid(True, which='both', linestyle='--', linewidth=0.5)

    individual_plots.append(fig)

plt.figure()
plt.plot([0, 1], [0, 1], color='orange', linestyle='--')

for i, fig in enumerate(individual_plots):
    ax = fig.axes[0]
    ax.lines[0].set_color(fold_colors[i])
    ax.lines[0].set_linestyle('--')
    plt.plot(ax.lines[0].get_xdata(), ax.lines[0].get_ydata(), color=fold_colors[i], linestyle='--', label=ax.lines[0].get_label())

plt.legend(loc='lower right')

# Customize the combined plot
plt.xlim([-0.02, 1.02])
plt.ylim([-0.02, 1.02])
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.title('ROC Curve for Logistic Regression (5-Fold Cross-Validation)')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
    
for i, a in enumerate(accuracy_logistic):
    # Print results
    print(f"Accuracy {i+1} : {a:.6f}")
print("Average Accuracy:", np.mean(accuracy_logistic))
print("Standard deviation of Accuracy:", np.std(accuracy_logistic))

for i, b in enumerate(precision_logistic):
    # Print results
    print(f"Precision {i+1} : {b:.6f}")
print("Average Precision:", np.mean(precision_logistic))
print("Standard deviation of Precision:", np.std(precision_logistic))
    
for i, c in enumerate(recall_logistic):
    # Print results    
    print(f"Recall {i+1} : {c:.6f}")
print("Average Recall:", np.mean(recall_logistic))
print("Standard deviation of Recall:", np.std(recall_logistic))
    
for i, d in enumerate(f1_logistic):
    # Print results    
    print(f"F-1 Score {i+1} : {d:.6f}")
print("Average F1-score:", np.mean(f1_logistic))
print("Standard deviation of F1-score:", np.std(f1_logistic))
```

**Metrics**

The changes in accuracy, precision, recall, and F-1 score across different folds are relatively small. The standard deviations for these metrics are also relatively small, indicating that the performance of the model is consistent across the dif-
